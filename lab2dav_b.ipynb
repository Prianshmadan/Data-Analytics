```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
sales = pd.read_csv('sales_data_sample.csv', encoding='ISO-8859-1')
sales.head()
```

```python
plt.figure(figsize=(12, 6))
product_line_years = sales.groupby(['YEAR_ID', 'ORDERLINENUMBER'])['SALES'].sum().unstack()
product_line_years.plot(kind='bar', stacked=True)
plt.title("Sales of Different Product Lines Each Year")
plt.xlabel("Year")
plt.ylabel("Sales Value")
plt.legend(title='Product Line')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sales.boxplot(column='SALES', by='YEAR_ID', vert=False)
plt.title("Boxplot of Total Sales Values Each Year")
plt.xlabel("Sales Value")
plt.ylabel("Year")
plt.suptitle("")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(sales['YEAR_ID'], sales['SALES'])
plt.title("Scatter Plot of Total Sales Values Each Year")
plt.xlabel("Year")
plt.ylabel("Sales Value")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist([sales[sales['YEAR_ID'] == year]['MONTH_ID'] for year in sales['YEAR_ID'].unique()],
         bins=12, stacked=True, label=sales['YEAR_ID'].unique())
plt.title("Histogram of Total Sales Monthwise and Yearwise")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend(title='Year')
plt.show()
```
